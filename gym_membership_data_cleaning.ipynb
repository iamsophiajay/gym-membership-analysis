```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('Customer_Subscription_And_Transaction_Details.csv')

# Display the first 5 rows
df.head()
```

```python
# List files in the current directory again
os.listdir()
```

```python
# Rename columns
df.rename(columns={
    'cust_id': 'member_id',
    'transaction_type': 'membership_status',
    'transaction_date': 'membership_date',
    'subscription_type': 'membership_plan',
    'subscription_price': 'membership_fee',
    'customer_gender': 'gender',
    'age_group': 'age_group',
    'customer_country': 'state',
    'referral_type': 'referral_source'
}, inplace=True)

# Check the updated column names
df.head()
```

```python
# Transform membership status
df['membership_status'] = df['membership_status'].replace({
    'initial': 'new',
    'UPGRADE': 'upgrade',
    'REDUCTION': 'downgrade',
    'CHURN': 'churn'
})

# Check the unique values in membership_status
df['membership_status'].value_counts()
```

```python
# Standardize membership plans
df['membership_plan'] = df['membership_plan'].str.upper()

# Check the unique values in membership_plan
df['membership_plan'].value_counts()
```

```python
# Standardize referral sources
df['referral_source'] = df['referral_source'].str.title()

# Check the unique values in referral_source
df['referral_source'].value_counts()
```

```python
# Check for missing values
df.isnull().sum()

# Drop rows with missing values (if necessary)
df.dropna(inplace=True)

# Alternatively, fill missing values (e.g., for referral_source)
df['referral_source'].fillna('Unknown', inplace=True)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Check the shape of the DataFrame after removing duplicates
df.shape
```

```python
# Export the cleaned data to a new CSV file
df.to_csv('Cleaned_Gym_Membership_Data.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Cleaned_Gym_Membership_Data.csv')

# Define the new membership fees
membership_fees = {
    'MAX': 12000,
    'BASIC': 2500,
    'PRO': 9000
}

# Update the membership_fee column based on the membership_plan
df['membership_fee'] = df['membership_plan'].map(membership_fees)

# Save the updated dataset to a new CSV file
df.to_csv('Updated_Gym_Membership_Data.csv', index=False)
```

```python
# Calculate total revenue
total_revenue = df['membership_fee'].sum()
print(f"Total Revenue: #{total_revenue:.2f}")
```

```python
# Count the number of unique customers
total_customers = df['member_id'].nunique()

# Count the number of customers with more than one transaction
retained_customers = df['member_id'].value_counts()[df['member_id'].value_counts() > 1].count()

# Calculate retention rate
retention_rate = (retained_customers / total_customers) * 100
print(f"Customer Retention Rate: {retention_rate:.2f}%")
```

```python
# Calculate the number of transactions (rows in the dataset)
total_transactions = df.shape[0]

# Calculate AOV
aov = total_revenue / total_transactions
print(f"Average Order Value (AOV): #{aov:.2f}")
```

```python
# Count the number of churned customers
churned_customers = df[df['membership_status'] == 'churn']['member_id'].nunique()

# Calculate churn rate
churn_rate = (churned_customers / total_customers) * 100
print(f"Churn Rate: {churn_rate:.2f}%")
```

```python
# Count the number of customers who upgraded
upgraded_customers = df[df['membership_status'] == 'upgrade']['member_id'].nunique()

# Calculate upgrade rate
upgrade_rate = (upgraded_customers / total_customers) * 100
print(f"Upgrade Rate: {upgrade_rate:.2f}%")
```

```python
# Count the number of customers who downgraded
downgraded_customers = df[df['membership_status'] == 'downgrade']['member_id'].nunique()

# Calculate downgrade rate
downgrade_rate = (downgraded_customers / total_customers) * 100
print(f"Downgrade Rate: {downgrade_rate:.2f}%")
```

```python
print("Summary of Key Metrics:")
print(f"Total Revenue: #{total_revenue:.2f}")
print(f"Customer Retention Rate: {retention_rate:.2f}%")
print(f"Average Order Value (AOV): #{aov:.2f}")
print(f"Churn Rate: {churn_rate:.2f}%")
print(f"Upgrade Rate: {upgrade_rate:.2f}%")
print(f"Downgrade Rate: {downgrade_rate:.2f}%")
```

```python
# Group by membership plan and calculate total revenue
revenue_by_plan = df.groupby('membership_plan')['membership_fee'].sum()

# Plotting
plt.figure(figsize=(8, 6))
revenue_by_plan.plot(kind='bar', color='blue')
plt.title('Revenue by Membership Plan')
plt.xlabel('Membership Plan')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=0)
plt.show()
```
